```python
def compute_score(lives_digits, score_digits, comp_digits_p1, comp_digits_p2, comp_digits_p3, comp_digits_p4 ,comp_digits_p5, comp_digits_l1):
    
    def mse(imageA, imageB):
        err = numpy.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
        err /= float(imageA.shape[0] * imageA.shape[1])
        return err

    def digit_comparison(in_digit,comp_digits):
        if numpy.sum(in_digit, axis=None) > 5000:
            errs={}
            for key, value in comp_digits.items():
                errs[key] = mse(in_digit,value)

            score_d = int(min(errs, key=errs.get)[3])
            score_is = 1
        else:
            score_d = 0
            score_is = 0
            
        return score_d, score_is
    
    score_d1, _ = digit_comparison(score_digits[0],comp_digits_p1)
    score_d2, _ = digit_comparison(score_digits[1],comp_digits_p2)
    score_d3, _ = digit_comparison(score_digits[2],comp_digits_p3)
    score_d4, _ = digit_comparison(score_digits[3],comp_digits_p4)
    score_d5, _ = digit_comparison(score_digits[4],comp_digits_p5)
    
    score = int(score_d5*10000+score_d4*1000+score_d3*100+score_d2*10+score_d1)
    
    lives_d1, _ = digit_comparison(lives_digits[0],comp_digits_l1)
    lives_d2, score_is = digit_comparison(lives_digits[1],comp_digits_l1)
    
    if score_is == 1:
        lives = int(lives_d1*10 + lives_d2)
    else:
        lives = int(lives_d1)
    
    return score, lives

#score, lives = compute_score(lives_digits, score_digits, comp_digits_p1, comp_digits_p2, comp_digits_p3, comp_digits_p4 ,comp_digits_p5, comp_digits_l1)
#benchmark("score, lives = compute_score(lives_digits, score_digits, comp_digits_p1, comp_digits_p2, comp_digits_p3, comp_digits_p4 ,comp_digits_p5, comp_digits_l1)")
#13421
```

```python
#This works, with just_game
def process_img(image):
    original_image = image
        
    # Convert BGR to HSV
    hsv = cv2.cvtColor(original_image, cv2.COLOR_BGR2HSV)
    
    # define range of blue color in HSV    
    hsv_lower = numpy.uint8([[[0, 50, 50]]])
    hsv_upper = numpy.uint8([[[359, 255, 255]]])
    
    # Threshold the HSV image to get only shapes
    threshold = cv2.inRange(hsv, hsv_lower, hsv_upper)
    
    processed_img = cv2.Canny(threshold, 30, 200)
    
    _, contours, _ = cv2.findContours(processed_img.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    perimeters  = []
    for c in contours:
        perimeters.append(cv2.arcLength(c,False))        
        
    contours_2 = []
    for i in range(len(contours)):
        if perimeters[i] < 20:
            pass
        else:
            contours_2.append(contours[i])        

    return contours_2

#contours = process_img(just_game)
#benchmark("contours = process_img(just_game)")
#498
```

```python
#This works, with just_game
def get_player(image):
    original_image = image
    height, width, _ = original_image.shape
    
    # Convert BGR to HSV
    hsv = cv2.medianBlur(cv2.cvtColor(original_image, cv2.COLOR_BGR2HSV),5)
 
    # define range in HSV    
    hsv_lower = numpy.array([15,50,50])
    hsv_upper = numpy.array([35,255,255])
    
    # Threshold the HSV image to get only shapes
    threshold = cv2.inRange(hsv, hsv_lower, hsv_upper)
    blanked = cv2.bitwise_and(original_image,original_image,mask = cv2.bitwise_not(threshold))
    
    processed_img = cv2.Canny(threshold, 30, 200)
        
    _, contours, _ = cv2.findContours(processed_img.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
        
    perimeters  = []
    for c in contours:
        perimeters.append(cv2.arcLength(c,False))
   
    areas  = []
    for c in contours:
        areas.append(cv2.contourArea(c))
        
    circles = []
    for c in contours:
        (x,y), radius = cv2.minEnclosingCircle(c)
        circles.append(math.pi*radius*radius)
        
    contours_2 = []
    for i in range(len(areas)):
        if circles[i] > 250:
            pass
        elif circles[i] < 25:
            pass
        elif areas[i] > 250:
            pass
        elif areas[i] < 25:
            pass
        elif perimeters[i] > 150:
            pass
        elif perimeters[i] < 20:
            pass
        else:
            contours_2.append(contours[i])
            
    filled = numpy.zeros((height,width,3))
    
    for c in contours_2:
        filled = cv2.fillPoly(filled, contours_2, color=(255,255,255))
 
    threshold_2 = cv2.inRange(filled, numpy.array([0,0,0]), numpy.array([0,0,255]))                 
    blanked_2 = cv2.bitwise_and(original_image,original_image,mask = cv2.bitwise_not(threshold_2))
       
    player_edge = cv2.Canny(blanked_2, 30, 200)
    
    _, players, _ = cv2.findContours(player_edge.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    
    player = sorted(players, key=cv2.contourArea, reverse=True)[0]

    moments = cv2.moments(player)
    centre_x = int(moments['m10']/moments['m00'])
    centre_y = int(moments['m01']/moments['m00'])

    return (centre_x, centre_y)

#player = get_player(just_game)
#benchmark("player = get_player(just_game)")
#113
```

```python
def non_para_direction(positions,height,width):

    i = len(positions)-1
    x_disp_pos = int(width - positions[i][0][0])
    x_disp_neg = int(positions[i][0][0])
    y_disp_pos = int(height - positions[i][0][1])
    y_disp_neg = int(positions[i][0][1])    

    x = []
    y = []
    t = []
    
    for p in positions:
        x.append(p[0][0])
        y.append(p[0][1])
        t.append((p[1] - datetime.datetime(1970,1,1)).total_seconds())
    
    #fx= interpolate.interp1d(t, x, kind='cubic', assume_sorted=True)
    #fy= interpolate.interp1d(t, y, kind='cubic', assume_sorted=True)
    
    fx= interpolate.UnivariateSpline(t, x, k=4)
    fy= interpolate.UnivariateSpline(t, y, k=4)
    
    t_new = numpy.linspace(t[0], t[-1], 15)
    x_new = fx(t_new)
    y_new = fy(t_new)
    
    x_delta = x_new[-1] - x_new[-2]
    y_delta = y_new[-1] - y_new[-2]
    
    tan_angle = numpy.arctan2([y_delta],[x_delta]) * 180 / numpy.pi
    
    angle = []
    if tan_angle >= -90:
        angle.append(90 + tan_angle[0])
    else:
        angle.append(450 + tan_angle[0])
    
    tan_thresh = numpy.arctan2([y_disp_neg,-y_disp_pos,-y_disp_pos,y_disp_neg],[x_disp_pos,x_disp_pos,-x_disp_neg,-x_disp_neg]) * 180 / numpy.pi
    
    for i in range(7):
        if angle[0]+(i+1)*45 < 360:
            angle.append(angle[0]+(i+1)*45)
        else:
            angle.append(angle[0]+(i+1)*45 - 360)

    thresholds = []    
    for t in tan_thresh:
        if t <= 90:
            thresholds.append(90 - t)
        else:
            thresholds.append(450 - t)

    line_ends = []
    for a in angle:
        
        if a <= thresholds[0]:
            line_end = (int(x_disp_neg + y_disp_neg*math.tan(math.radians(a))),0)
        elif a <= thresholds[1]:
            if a <= 90:
                line_end = (width,int(y_disp_neg-x_disp_pos*math.tan(math.radians(90-a))))
            else:
                line_end = (width,int(y_disp_neg+x_disp_pos*math.tan(math.radians(a-90))))
        elif a <= thresholds[2]:
            if a <= 180:
                line_end = (int(x_disp_neg+y_disp_pos*math.tan(math.radians(180-a))),height)
            else:
                line_end = (int(x_disp_neg-y_disp_pos*math.tan(math.radians(a-180))),height)
        elif a <= thresholds[3]:
            if a <= 270:
                line_end = (0,int(y_disp_neg+x_disp_neg*math.tan(math.radians(270-a))))
            else:
                line_end = (0,int(y_disp_neg-x_disp_neg*math.tan(math.radians(a-270))))
        else:
            line_end = (int(x_disp_neg-y_disp_neg*math.tan(math.radians(360-a))),0)
        
        line_ends.append(line_end)

    return line_ends
```

```python
def get_x_lines(player,line_ends):
    x_lines = []
    x1 = player
    for le in line_ends:
        x2 = le
        A = x1[1]-x2[1]
        B = x2[0]-x1[0]
        C = B*x2[1]+A*x2[0]
        x_lines.append([A,B,C])
    
    return x_lines

#x_lines = get_x_lines((391,50),[(391,500),(640,400)])
```

```python
def get_intersections(x_lines,contours):
    intersections = []
    for x_line in x_lines:
        for contour in contours:
            c_lines = []
            for i in range(len(contour)):
                if i == 0:
                    c2 = contour[i][0]
                else:
                    c1 = contour[i][0]
                    A = c1[1]-c2[1]
                    B = c2[0]-c1[0]
                    C = B*c2[1]+A*c2[0]
                    c_lines.append([c1,c2,A,B,C])
                    c2 = contour[i][0]

            for i in range(len(c_lines)):
                det = x_line[0]*c_lines[i][3] - c_lines[i][2]*x_line[1]
                c_lines[i].append(det)

            for i in range(len(c_lines)):
                if c_lines[i][5] == 0:
                    intersect = 0
                    x = 0
                    y = 0
                else:
                    intersect = 1
                    x = (c_lines[i][3]*x_line[2] - x_line[1]*c_lines[i][4]) / c_lines[i][5]
                    y = (x_line[0]*c_lines[i][4] - c_lines[i][2]*x_line[2]) / c_lines[i][5]

                c_lines[i].append(intersect)
                c_lines[i].append(x)
                c_lines[i].append(y)

            inter = []
            for c in c_lines:
                if c[5] == 0:
                    pass
                elif c[-2] > max(c[0][0],c[1][0]):
                    pass
                elif c[-2] < min(c[0][0],c[1][0]):
                    pass
                elif c[-1] > max(c[0][1],c[1][1]):
                    pass
                elif c[-1] < min(c[0][1],c[1][1]):
                    pass
                else:
                    inter.append((c[-2],c[-1]))

            intersect = list(set(inter))
            if len(intersect) != 0:
                for i in intersect:
                    intersections.append(i)

    return intersections

#intersections = get_intersections(x_lines,contours)
```
